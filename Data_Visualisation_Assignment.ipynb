```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
mpl.rcParams['text.color'] = 'white'
mpl.rcParams['axes.labelcolor'] = 'white'
mpl.rcParams['xtick.color'] = 'white'
mpl.rcParams['ytick.color'] = 'white'
```

```python
features_df = pd.read_csv('/content/drive/MyDrive/Music DataSet/Features.csv')
```

```python
print(features_df.head())
```

```python
list(features_df.columns)
```

```python
streams_df = pd.read_csv('/content/drive/MyDrive/Music DataSet/Streams.csv')
```

```python
merged_df = pd.merge(streams_df, features_df, left_on='Song', right_on = "name")
```

```python
sb.scatterplot(x='danceability', y='Streams (Billions)', data=merged_df)
plt.title('Danceability vs Streams')
plt.xlabel('Danceability')
plt.ylabel('Streams')
plt.show()
```

```python
sb.scatterplot(x='loudness', y='danceability', data=features_df)
plt.title('Comparison of Danceability by loudness')
plt.show()
```

```python
sb.lineplot(x='energy', y='danceability', data=features_df)
plt.title('Distribution of danceability by energy')
plt.show()
```

```python
sb.scatterplot(x='speechiness', y='danceability', data=features_df)
plt.title('Distribution of danceability by speechiness')
plt.show()
```

```python
sb.scatterplot(x='duration', y='danceability', data=features_df)
plt.title('Distribution of danceability by duration of the song')
plt.xlabel('duration in minutes')
plt.show()
```
